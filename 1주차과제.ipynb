#인공지능 제품 또는 서비스(4가지)

## 1. 네이버의 '파파고' (언어 분야)
* 파파고(Papago)는 에스페란토어로 '앵무새'란 뜻으로, 2017년 8월에 정식 출시된 인공지능 번역 서비스이다.
* 인공신경망 기반 N2MT 기술을 사용하여, 문장의 전체의 맥락을 먼저 이해하고 구성 요소들을 번역하는 방식을 활용한다.

## 2. 네이버의 '클로바' (음성 분야)
* 클로바(Clova)는 2017년 5월 출시되었다.
* 2019년 7월 새롭게 업그레이드된 HDTS 방식을 사용하는데, 이것은 딥러닝(DNN)과 파형 접합 합성(UTS)을 결합한 하이브리드 음성합성(Hybrid Dnn Text-to-Speech) 기술이다.
* 인공지능 음성인식 기술로는 이외에도 구글 홈미니, 애플 시리, 아마존 알렉사, SK 누구 등 다양한 제품과 서비스가 있다.

## 3. 두잉랩의 '푸드 렌즈' (이미지 분야)
* 푸드 렌즈(Food Lens) 앱이 인식하는 음식은 한식·중식·양식·가공식품 등 총 6000개 정도이고, 인식 정확도는 90%가 넘는다고 한다.
* 사진 한 장으로 정밀한 열량 계산과 식습관 분석을 할 수 있다고 하는데, 딥러닝 이미지 처리 기술을 사용하며, 푸드렌즈가 딥러닝을 통해 지금까지 데이터를 학습한 사진은 100만장이 넘는다고 한다.

## 4. 테슬라의 자율주행차량 '모델3'(자율주행차 분야)
* 일론 머스크가 이끄는 테슬라의 자율주행차량 '모델3'는 오토 파일럿(Autopilot) 자율주행 시스템을 기반으로 다양한 센서들과 통신 기술들이 합쳐져 만들어졌으며, 현대, 구글, GM 등 다른 회사들의 자율주행 차량과는 현격한 기술 차이를 보이며 앞서 가고 있다.
* 참고로, 2016년부터 국제자동차기술자협회(SAE International)에서 자율주행차량을 분류한 단계가 글로벌 기준으로 통용되고 있는데 ‘레벨0’에서 ‘레벨5’까지 6단계로 나뉜다. 레벨 0은 전방 충돌방지 보조(FCA), 후측방 충돌경고(BCW) 등, 레벨 1은 차로 유지보조(LFA), 스마트 크루즈 컨트롤(SCC) 등, 레벨 2는 특정 주행모드에서 시스템이 조향 및 감·가속을 모두 수행하는 고속도로 주행 보조(HDA)로 분류된다. 레벨 2까지는 시스템이 주행을 돕지만 레벨 3부터는 특정 모드에서 시스템이 주행을 수행하는데, 자동차 스스로 차선을 변경하고 앞차를 추월하거나 장애물을 피할 수 있게 된다. 레벨 3는 차선 감지를 통해 운전자가 핸들 조작하지 않아도 자동으로 운전방향을 잡아주지만 위험예상시 사용자에게 조작권을 넘기며(고속도로 자율주행), 레벨 4는 위험 상황도 스스로 처리하고 도심에서도 자율주행이 가능한 조건부 완전 자율주행, 레벨 5는 지역 제한이 없는 완전 자율주행이다.

* 올해 7월부터 우리나라에서 레벨3 자율주행차랑의 판매가 가능해지며 이번 성공 여부가 향후 레벨4, 5로 가기 위한 중요한 계기가 될 것이라고 한다.
